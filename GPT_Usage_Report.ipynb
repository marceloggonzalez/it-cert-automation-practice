## Librerias

```python
#!pip install --upgrade gspread google-auth google-auth-oauthlib google-auth-httplib2
from google.colab import auth
import gspread
from google.auth import default
import pandas as pd
from matplotlib import pyplot as plt
from PyPDF2 import PdfMerger

# Autenticación
auth.authenticate_user()

# Obtener credenciales predeterminadas
creds, _ = default()

# Autorizar el acceso a Google Sheets
gc = gspread.authorize(creds)
```

```python
peso_actividad = 0.7
peso_mensajes = 0.3
google_sheet_url_low_usage = 'https://docs.google.com/spreadsheets/d/1ZZSFHWHQFvggn6SOvXtND4fh6sbAmXlCnZtWgCSeItA/edit?usp=drive_link'
google_sheet_url_top_usage = 'https://docs.google.com/spreadsheets/d/1ahnQcDbOjGw7icmyjVrO8EIMPVKq9s8PPNtJB6LqMEs/edit?usp=drive_link'
google_sheet_url_gpts_info = 'https://docs.google.com/spreadsheets/d/1EQjt6t2Ml8Gq71CWrO9k02ulEO9EeZczQ4wowABa5l4/edit?gid=1849472582#gid=1849472582&range=A1'
```

## Inputs

### Low Usage

```python
# Abre la hoja de cálculo por URL
low_usage_sheet = gc.open_by_url(google_sheet_url_low_usage)

# Selecciona una hoja específica
low_usage_worksheet = low_usage_sheet.worksheet('Low Usage')

# Obtener todos los valores de la hoja
low_usage_data = low_usage_worksheet.get_all_values()
print(low_usage_data)
```

### Top usage

```python
# Abre la hoja de cálculo por URL
top_usage_sheet = gc.open_by_url(google_sheet_url_top_usage)

# Selecciona una hoja específica
top_usage_worksheet = top_usage_sheet.worksheet('Top Usage')

# Obtener todos los valores de la hoja
top_usage_data = top_usage_worksheet.get_all_values()
print(top_usage_data)
```

## GPTs Info

```python
# Abre la hoja de cálculo por URL
gpts_info_sheet = gc.open_by_url(google_sheet_url_gpts_info)

# Selecciona una hoja específica
gpts_info_worksheet = gpts_info_sheet.worksheet('GPTs')

# Obtener todos los valores de la hoja
gpts_info_data = gpts_info_worksheet.get_all_values()
print(gpts_info_data)
```

## Pre Process

__Low Usage__

```python
df_lu = pd.DataFrame(low_usage_data)
df_lu = df_lu.rename(columns=df_lu.iloc[0]).drop(df_lu.index[0]).reset_index(drop=True)
df_lu = df_lu[df_lu.domain != 'C. MONITORING'].replace("N/A", 0)
```

```python
columnas = ['days_active_last_15d', 'days_active_last_30d', 'days_active_last_60d', 'All_messages_60d']
df_lu[columnas] = df_lu[columnas].apply(pd.to_numeric, errors='coerce')
df_lu['tasa_actividad'] = round((df_lu['days_active_last_60d']/60)*100, 2)
df_lu['msjs_por_dia'] = round((df_lu['All_messages_60d']/60)*100, 2)
df_lu['combined_index'] = round((df_lu['tasa_actividad'] * peso_actividad) + (df_lu['msjs_por_dia'] * peso_mensajes), 2)
```

```python
df_lu_sorted = df_lu.sort_values(by=['domain', 'combined_index'])

# Luego agrupamos por 'domain' y seleccionamos las primeras 3 filas de cada grupo
df_lu_result = df_lu_sorted.groupby('domain').head(20)
```

__Top Usage__

```python
df_tu = pd.DataFrame(top_usage_data)
df_tu = df_tu.rename(columns=df_tu.iloc[0]).drop(df_tu.index[0]).reset_index(drop=True)

df_tu = df_tu.replace("N/A", 0)
df_tu = df_tu[df_tu.domain != 'C. MONITORING']
```

```python
columnas = ['days_active_last_60d',	'All_messages_60d',	'GPTs_created_60d',	'Countries_using_your_GPTs_60d',	'Domains_using_your_GPTs_60d',	'Users_using_your_GPTs_60d',	'Messages_using_your_GPTs_60d']
df_tu[columnas] = df_tu[columnas].apply(pd.to_numeric, errors='coerce')
```

```python
df_tu['tasa_actividad'] = round((df_tu['days_active_last_60d']/60)*100, 2)
df_tu['msjs_por_dia'] = round((df_tu['All_messages_60d']/60)*100, 2)
df_tu['combined_index'] = round((df_tu['tasa_actividad'] * peso_actividad) + (df_tu['msjs_por_dia'] * peso_mensajes), 2)
```

__GPTs Info__

```python
gpts_info = pd.DataFrame(gpts_info_data)
gpts_info = gpts_info.rename(columns=gpts_info.iloc[0]).drop(gpts_info.index[0]).reset_index(drop=True)
```
